# Day1_Transformers

```python
from transformers import pipeline
```

#### Sentiment Analysis
The model will label output as (negative or positive) and score in the range of 0-100%

```python
classifier = pipeline("sentiment-analysis")
classifier(["Today we're in protesting for the new governement creation on parliament.",
           "Criminals who run from the jail are doing criminal activities"])
```

```python
classifier = pipeline("zero-shot-classification")
classifier("This is a course about the transformers", candidate_labels = ['education', 'Politics', 'Buiseness'])
```

#### Text Generation

```python
generator = pipeline("text-generation")
generator(
    "In this course we will teach you how to "
)
```

#### distilgpt2 Text Generation

```python
generators = pipeline("text-generation", model='distilgpt2')
generators("In this course, we will teach you how to",
           max_length=2,
           num_return_sequences=2)
```

```python
unmasker = pipeline('fill-mask')
unmasker("This course will teach you all about <mask> models", top_k=2)
```

```python
ner = pipeline("ner", grouped_entities=True)
ner("My name is shiva chaudhary and i work at Hugging Face in Brooklyn")
```

```python
ner = pipeline("ner", aggregation_strategy='simple')
ner("My name is shiva chaudhary and i work at Facebook")
```

```python
summarizer = pipeline("summarization")
summarizer("Google was founded on September 4, 1998, by American computer scientists Larry Page and Sergey Brin. Together, they own about 14% of its publicly listed shares and control 56% of its stockholder voting power through super-voting stock. The company went public via an initial public offering (IPO) in 2004.")
```

```python
question_answering = pipeline("question-answering")
question_answering(question="Where do I work ?",
                  context = "My name is shiva and i work in hugging face at brooklyn")
```

```python
gpt2 = pipeline(task='text-generation', model="openai-community/gpt2")
gpt2("I am shiva chaudhary", num_return_sequences=2)
```

```python
results = gpt2("What if AI destroy ", num_return_sequences=2, max_new_tokens=100)
```

```python
for result in results:
    print(result['generated_text'])
```
