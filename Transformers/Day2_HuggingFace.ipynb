# Grammar Checker


```python
from transformers import pipeline

pipe = pipeline("text-classification", model="abdulmatinomotoso/English_Grammar_Checker")
```

```python
# load model directly
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained("abdulmatinomotoso/English_Grammar_Checker")

model = AutoModelForSequenceClassification.from_pretrained("abdulmatinomotoso/English_Grammar_Checker")
```

The label 0 means that the grammar is incorrect with the confidence score.

```python
print(pipe("He eat pizza every day."))
```

The label 1 means that the grammar is correct with the confidence score : 0.99

# Text Classification: Dynamic category assignment

Dynamically assigns categories based on content.

Request : I want to know more about your pricing plans

categories :
sales -> High Confidence (True)

marketing -> Moderate Confidence (False)

Support -> Low Confidence (False)


```python
classifier = pipeline(
    task = "zero-shot-classification",
    model = "facebook/bart-large-mnli"
)

text = "Hey, DataCamp; We would like to feature your course in our newsletter!"
categories = ['marketing', 'sales', 'support']

output = classifier(text, categories)

print(f"Top Label: {output['labels'][0]} with score : {output['scores'][0]}")
```

# Generation of Text
- Extractive
  - are the the text where it selects keys sentences from texts.
  - it is efficient and does not need extra computaitonal powes, fewer resources
  - but it lacks flexibility
  for example: law files, reports

- Abstractive
  - Generates new, rephrased text
  - clearer and mode readable
  - Requires more resources and processing
  for example: newspaper



```python
# Load the abstraction summarization pipeline

summarizer = pipeline("summarization", model="sshleifer/distilbart-cnn-12-6")

text = "max_length is an attribute used primarily with Django's CharField and SlugField to specify the maximum allowed number of characters in that field. Example: In this example, the name field can store up to 100 characters. If a user tries to input more than 100 characters, Django will raise a ValidationError . "
summary_text = summarizer(text)
print(summary_text)
```
